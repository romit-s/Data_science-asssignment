```python
import numpy as np 
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("breastcancer.csv")
df.head()
```

```python
df_label=df["class"]
df_features=df.drop("id",1)
df_features=df_features.drop("class",1)
df_features.replace('?',-99999,inplace=True)
df_features.head()
```

```python
label=[]
for lab in df_label:
    if lab==2:
        label.append([1,0])
    elif lab==4:
        label.append([0,1])
```

```python
features=np.array(df_features)
labels=np.array(label)
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
model=Sequential()
model.add(Dense(500,input_dim=9,activation='sigmoid'))
model.add(Dense(100,activation='tanh'))
model.add(Dense(2,activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x_train,y_train, epochs=1000, batch_size=50, validation_data=(x_test,y_test))
```

```python
test=np.array([[5,1,1,1,2,1,3,1,1]])
result=model.predict_classes(test)
print(result)
if result==0:
    print("benign")
else:
    print("malignant")
```
